```python
import pandas as pd
import matplotlib.pyplot as plt
# scatter plot for winning number 1
# Load the CSV file
file_path = 'sorted_data.csv'  # Replace with your CSV file path
column_name = 'winning_number_1'    # Replace with the name of the column you want to plot

# Read the CSV file
data = pd.read_csv(file_path)

# Extract the integers from the specified column
values = data[column_name].dropna().astype(int)

# Create a scatter plot
plt.scatter(range(len(values)), values)
plt.title('Scatter Plot of ' + column_name)
plt.xlabel('Index')
plt.ylabel(column_name)
plt.show()
```

```python
# tried to make a bar chart to show the counter for all the numbers, but keep getting errors
# Load the CSV file
# Replace 'your_file.csv' with the path to your CSV file
data = pd.read_csv('sorted_data.csv', header=None)

# Flatten the DataFrame to a single series and count occurrences
counts = data.stack().value_counts().reindex(range(1, 71), fill_value=0)

# Create the bar graph
plt.figure(figsize=(12, 6))
plt.bar(counts.index, counts.values)
plt.xlabel('Numbers (1-70)')
plt.ylabel('Count')
plt.title('Count of Numbers from CSV')
plt.xticks(range(1, 71))  # Set x-ticks to show all numbers from 1 to 70
plt.yticks(range(0, 2501, 250))  # Set y-ticks to show counts up to 2500
plt.grid(axis='y')

# Show the plot
plt.show()
```
